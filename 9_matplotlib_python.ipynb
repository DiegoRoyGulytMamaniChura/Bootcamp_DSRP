## Parte 2:

```python
#importar matplotlib + pandas
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/1 clases python/universidad_data.csv")
```

```python
data.info()
```

```python
data.head(5)
```

### 1. Grafica de Barras: Costo promedio por universidad

```python
#Primero hacemos nuestro Groupby.
promedio = data.groupby('Universidad')['Costo_por_ao_USD'].mean().round(2)

print(promedio,"\n")

#Segundo separar el index y los values
universidad = promedio.index
valores = promedio.values

print("\n univ: ", universidad)
print("\n val: ",valores)
```

```python
#Grafico de Barras:
#Figura:
plt.figure(figsize=(6,3))

#Plot
plt.bar(universidad,promedio)

#Titulo:
plt.title("Universidades: Costo por ao")

#Etiqueta de los ejes:
plt.xlabel("Universidad")
plt.ylabel("Promedio de Costo")

plt.xticks(rotation=90)
plt.show()
```

## EJERCICIO: Hagan lo mismo, indiquen la Nacionalidad y el conteo de la nacionalidad

```python
nacionalidad=data.groupby('Nacionalidad')['Nacionalidad'].count()

print(nacionalidad,"\n")

#Segundo
na = nacionalidad.index
cantidad = nacionalidad.values

print("\n univ: ", na)
print("\n val: ", cantidad)
```

```python
#Figura:
plt.figure(figsize=(6,3)) #Alto y Ancho esta medido en pulgadas y esto se
plt.bar(na, cantidad,color=('red','blue','green'))

#Titulo:
plt.title("Count de nacionalidades")

#Etiqueta de los ejes:
plt.xlabel("Nacionalidad")
plt.ylabel("Count")

plt.xticks(rotation=50)
plt.show()
```

## Grafico de Histograma

```python
#Histograma: Distribucion de la edad de los estudiantes.
plt.figure(figsize=(6,3))
plt.hist(data['Edad'],color='skyblue',edgecolor='black', bins=20)

plt.title('Histograma de Edad')
plt.xlabel('Edad')
plt.ylabel('Numero de Estudiantes')
plt.show()
```

```python
#Histograma: Distribucion de los costos:
plt.figure(figsize=(6,3))
plt.hist(data['Costo_por_ao_USD'],color='skyblue',edgecolor='black', bins=10)

plt.title('Histograma de Costo')
plt.xlabel('Costo')
plt.ylabel('Cantidad de Costos')
plt.show()
```

## Grafico de Pie (Circulo)

```python
#Grafico de Pie (Circulo/Pastel): Distribucion de estudiantes por universidad:

#Primero contamos a los estudiantes:
conteo = data.groupby('Universidad')['Universidad'].count()
print(conteo,"\n")

universidad = conteo.index
conteo_valores = conteo.values

print(universidad)
print(conteo_valores)
```

```python
#Grafico de Pie

plt.pie(conteo_valores, labels=universidad, autopct='%1.1f%%')
plt.title('Distribucin de Estudiantes por Univesidad')
plt.show()
```

## EJERCICIO: Haga lo mismo, pero usando la variable Sexo

```python
#Liz Reyes:
conteo= data.groupby('Sexo')['Sexo'].count()
print(conteo,'\n')

#segundo paso
sexo= conteo.index
cont_valores= conteo.values
print(sexo)
print(cont_valores)
```

```python
plt.pie(cont_valores, labels=sexo, autopct='%1.1f%%')
plt.title('Sexo')
plt.show
```

```python
#Grafica de Box Plot:

#plt.figure()
plt.boxplot(data['Edad'])
plt.title('Distribucin de Edades de los Estudiantes')
plt.ylabel('Edad')
plt.xticks([1],['Estudiantes'])
plt.show()

"""
xticks es una fucnion para estables las ubicaciones y etiquetas del eje X
"""
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ventas = pd.read_csv("/content/drive/MyDrive/1 clases python/data_ventas.csv")
```

```python
ventas.info()
```

### 1. Grafico de Lineas

```python
#Grafico de Lineas usando Year y Ventas_1

plt.figure(figsize=(8,4))
plt.plot(ventas['Year'], ventas['Ventas_1'], color='#00FF00', marker='o', linestyle='--')
plt.title('Ventas anuales')
plt.xlabel('Ao')
plt.ylabel('Ventas 1')
plt.grid(True, linewidth=1, linestyle='--')
plt.show()
```

```python
#Grafico de dispersion
```

```python
#Grafico de dispersion usando Year y Ventas_2
plt.figure(figsize=(8,6))
plt.scatter(ventas['Ventas_1'],ventas['Ventas_2'], color='#FE2EC8')
plt.title('Dispersion')
plt.xlabel('Venta 1')
plt.ylabel('Ventas 2')
plt.show()
```

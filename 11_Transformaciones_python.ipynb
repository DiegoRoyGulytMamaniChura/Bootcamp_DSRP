# TRANSFORMACIONES

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/1 Bootcamp_Python_DSRP/5_Quinta_Clase/Ejercicio 2.xlsx")

base_nueva = df

base_nueva.head(20)
```

```python
base_nueva.info()
```

##  Limpieza y preparacion de datos:

En este paso, rellenamos lo valores faltanse (NaN)
** FECHA DE ACTIVACION

```python
# 1 LIMPIEZA Y PREPRACION DE DATOS:
#Rellenear los valores falantes en la columna 'FECHA ACTIVACION' con una fecha ficticia '1900-01-01'
#fillna -> nos va a permirt rellenar la data faltante
#timesstamp -> nos permite trabajar con fechas.

base_nueva['VENDEDOR'].fillna(pd.Timestamp('1900-01-01'), inplace=True)

base_nueva.head(2)
```

```python
base_nueva['DEPARTAMENTO'].fillna('DESCONOCIDO',inplace=True)

base_nueva.head(5)
```

## ELIMINACION DE DUPLICADOS:


```python
#Eliminacion de duplicados:
#drop_duplicates ->elimina filas totalmente duplicadas

base_nueva.drop_duplicates(inplace=True)

base_nueva.info()
```

```python
## TRANSFORMACION DE DATOS:
base_nueva.head(2)
```

```python
#DIFERENCIA ENTRE FECHA DE VENTA Y FECHA DE ACTIVACION:
base_nueva['DURACION_DIAS'] = (base_nueva['FECHA ACTIVACION']-base_nueva['FECHA VENTA']).dt.days

base_nueva.head(10)
```

```python
#reemplazo condicional
#1900-01-01 -> no sea activo o no disponible

base_nueva.loc[base_nueva['FECHA ACTIVACION'] == pd.Timestamp('1900-01-01'),'DURACION_DIAS']=-1

base_nueva.head(3)
```

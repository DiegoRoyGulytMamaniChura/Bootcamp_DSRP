# SEABORN - PARTE 3

## RELATION

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
propinas = sns.load_dataset("tips")
propinas.head(10)
```

```python
sns.scatterplot(
    data=propinas,
    x='total_bill',
    y='tip')
plt.show()
```

```python
sns.scatterplot(    data=propinas,    x='total_bill',    y='tip', hue='day',style='time',size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Marcadores:
marker={"Lunch":"D","Dinner":"s"}
sns.scatterplot(    data=propinas,    x='total_bill',    y='tip', hue='day',style='time',size='size',markers=marker)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Relplot:

sns.relplot(data=propinas, x='total_bill', y='tip',hue='day',style='time',size='size', col='time')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Relplot:

sns.relplot(data=propinas, x='total_bill', y='tip',hue='day',style='time',size='size', col='time')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Relplot:

sns.relplot(data=propinas, x='total_bill', y='tip',hue='day',style='time',size='size', col='day')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```


## JOINPLOT + PAIRPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

propinas = sns.load_dataset("tips")
propinas.head(2)
```

```python
#Jointplot
plt.figure(figsize=(8,6))
sns.jointplot(data=propinas, x='total_bill',y='tip')
plt.show()
```

```python
#Jointplot
plt.figure(figsize=(8,6))
sns.jointplot(data=propinas, x='total_bill',y='tip', hue='sex', kind='hist')
plt.show()
```

```python
#Jointplot + marginals
plt.figure(figsize=(8,6))
sns.jointplot(data=propinas, x='total_bill',y='tip', hue='sex', kind='hist',
              marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
#PAIRPLOT
sns.pairplot(data=propinas)
plt.show()
```

```python
#PAIRPLOT
sns.pairplot(data=propinas, hue='time')
plt.show()
```

## HEATMAP

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

propinas = sns.load_dataset("tips")
propinas.head(2)
```

```python
#Pandas nos da la correlacion:

propinas.corr()
```

```python
#Heatmap

sns.heatmap(propinas.corr())
plt.show()
```

```python
#Contexto:

sns.heatmap(propinas.corr(),annot=True,cmap='coolwarm', linewidths=5,cbar=True)
plt.show()
```

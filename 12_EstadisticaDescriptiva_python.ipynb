# ESTADISTICA DESCRIPTIVA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Cargar el DataFrame
df = pd.read_excel("/content/drive/MyDrive/1 Bootcamp_Python_DSRP/5_Quinta_Clase/Ejercicio 2.xlsx")
df.head(5)
```

```python
df['DURACION_DIAS'] = (df['FECHA ACTIVACION']-df['FECHA VENTA']).dt.days

df.head(2)
```

#Medidas de tendencia central:

*  Media: el promedio de los datos
*  Mediana: el valor central de un conjunto de datos
*  Moda:es el que aparece con más frecuencia.

```python
#Medidas de Tendencia Central:

media = df['FLUJO MES VENTA'].mean() #Promedio

print(media)

mediana = df['FLUJO MES VENTA'].median()
print(mediana)

moda = df['FLUJO MES VENTA'].mode()[0]
print(moda)
```

# Medidas de dispersion:

*  Varianza: Es el promedio de los cuadrados de las difrencia media.
*  Desviacion Standar: raiz cuadrada de la varianza


```python
varianza = df['DURACION_DIAS'].var()
std = df['DURACION_DIAS'].std()

print(varianza)
print(std)
```

```python
#Visualizacion:
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(14, 6))
sns.histplot(df['FLUJO MES VENTA'], kde=True)
plt.title('Histograma de Duración')
plt.xlabel('Duración')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x=df['FLUJO MES VENTA'])
plt.title('Boxplot de Duración')
plt.xlabel('Duración')
plt.show()
```

## EDA APLICADO A CULQI

```python
#Cargamos nuestra data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

culqi_df = pd.read_excel("/content/drive/MyDrive/1 clases python/Ejercicio 2_.xlsx")

culqi_df.head()

cdf = culqi_df.dropna()
```

```python
culqi_df.info()
```

Distribución de Variables:

*  ¿Cuáles son los canales de venta más comunes?
*  ¿Qué productos se venden más?
*  ¿Cómo se distribuyen las ventas entre los diferentes departamentos?
*  ¿Hay algún vendedor específico que tenga un rendimiento notablemente alto o bajo?

```python
#realizamos una tabla para contar canales de venta más comunes
df = cdf['CANAL VENTA'].value_counts(sort=True).reset_index(name='count')


sns.barplot(data=df,
            x='index',
            y='count')
```

```python
#realizamos una tabla para contar canales de venta más comunes
df = cdf['PRODUCTO'].value_counts(sort=True).reset_index(name='count')


sns.barplot(data=df,
            y='index',
            x='count')
```

```python
#realizamos una tabla para contar canales de venta más comunes
df = cdf['DEPARTAMENTO'].value_counts(sort=True).reset_index(name='count')


sns.barplot(data=df,
            y='index',
            x='count')
```

```python
#realizamos una tabla para contar canales de venta más comunes
df = cdf['VENDEDOR'].value_counts(sort=True).reset_index(name='count').head(5)


sns.barplot(data=df,
            y='index',
            x='count')
```

```python
#realizamos una tabla para contar canales de venta más comunes
df = cdf['VENDEDOR'].value_counts(sort=True).reset_index(name='count').tail(5)


sns.barplot(data=df,
            y='index',
            x='count')
```
